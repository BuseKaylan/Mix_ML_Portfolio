```python
plt.figure(figsize=(20,8))
plt.bar(df['Make'], df['Fuel_Tank_Capacity(L)'], color = 'grey', edgecolor = 'black')
```

```python
plt.figure(figsize = (4,3))
plt.title('Power vs. Torque of Volkswagen vs. Toyota')
plt.xlabel('Power (BHP)')
plt.ylabel('Torque (Nm)')
plt.scatter(df[df['Make'] == 'Volkswagen']['Power(BHP)'], df[df['Make'] == 'Volkswagen']['Torque(Nm)'], label = 'Volkswagen',marker = 'v', color = 'orange', s = 10)
plt.scatter(df[df['Make'] == 'Toyota']['Power(BHP)'], df[df['Make'] == 'Toyota']['Torque(Nm)'], label = 'Toyota', marker = 'o', color = 'blue', s = 10)
plt.legend(bbox_to_anchor = (1.0, 1.0), shadow = True, fontsize = 'medium')
plt.show()
```

```python
df.info()
```

```python
df['Price'] = df['Price'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df.info()
```

```python
plt.figure(figsize=(10,5))
plt.title('Price of the Cars')
plt.bar(df['Make'], df['Price'], color = 'darkblue')
plt.xticks(rotation = 45)
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title('Price Respect to Transmission Type')
plt.bar(df['Transmission_Type'], df['Price'], color = 'black')

plt.subplot(1,2,2)
plt.title('Price Respect to Fuel Type')
plt.bar(df['Fuel_Type'], df['Price'], color = 'grey')
```
